```python
from rosreestr2coord import Area
import pandas as pd
import json
from geojson import Polygon
import geojson
```

```python
cad_nums = ["42:07:0111003:1945","42:07:0111001:325","42:07:0111003:1944","70:17:0000009:466","70:17:0000009:468","70:17:0000009:444","70:17:0000009:467","70:17:0000009:461","70:17:0000009:1070","70:17:0000009:291","70:17:0000009:1025","70:17:0000009:1026","70:17:0000009:1028","70:17:0000009:1029","70:17:0000009:459","70:17:0000009:460","70:17:0000013:254","70:17:0000013:255","70:17:0000013:33","70:13:0100012:302","70:13:0101002:97","70:17:0000021:4","70:00:0000000:98", "70:17:0000009:1349", "70:17:0000009:1078"]
cad_nums_rent = ["70:02:0200046:1269","70:17:0000009:46","70:17:0000009:1071","70:17:0000000:259","70:17:0000015:157","70:17:0000009:0469" , "70:13:0100016:417", "70:13:0100016:418", "70:17:0000013:983", "70:17:0000013:989", "70:17:0000015:4060"]
```

```python
areas = [Area(num) for num in cad_nums]
areas_rent = [Area(num) for num in cad_nums_rent]
```

```python
names = [area.code for area in areas if  area.get_coord() != []]
coords= [area.get_coord() for area in areas if area.get_coord() != []]
ids = [i for i in range(len(areas))]
```

```python
names_rent = [area.code for area in areas_rent if  area.get_coord() != []]
coords_rent = [area.get_coord() for area in areas_rent if area.get_coord() != []]
```

```python
features_list = []
```

```python
ids = 0
for i,j in zip(coords,names):
    ids = ids + 1
    feature={
    "type": "Feature",
    "id": ids ,
    "geometry": 
    {"type": "Polygon",
    "coordinates": i },
    "properties": 
    {"description": j,
    "fill": "#FF0000",
    "fill-opacity": 0.55,
    "stroke": "#000000",
    "stroke-width": "1",
    "stroke-opacity": 1.0}}
    
    features_list.append(feature)
```

```python
ids = len(features_list)
for i,j in zip(coords_rent,names_rent):
    ids = ids + 1
    feature={
    "type": "Feature",
    "id": ids ,
    "geometry": 
    {"type": "Polygon",
    "coordinates": i },
    "properties": 
    {"description": j,
    "fill": "#483D8B",
    "fill-opacity": 0.9,
    "stroke": "#000000",
    "stroke-width": "1",
    "stroke-opacity": 1.0}}
    
    features_list.append(feature)
```

```python
geo = {
    "type" : "FeatureCollection",
    "crs" : { "type" : "name",
              "properties" : {
                  "name" : "EPSG:4326"
                  }
              },
    "features" : None
    }
```

```python
geo['features'] = features_list
```

```python
with open('result2.json', 'w') as fp:
     json.dump(geo, fp)
```
