```python
areas = [Area(num) for num in cad_nums]
df = pd.DataFrame([area.get_attrs() for area in areas if area.get_coord() !=[]])
df['geometry'] = [area.get_coord() for area in areas if  area.get_coord() != []]

```

```python
features_list = []

geo = {
    "type" : "FeatureCollection",
    "crs" : { "type" : "name",
              "properties" : {
                  "name" : "EPSG:4326"
                  }
              },
    "features" : None
    }
```

```python
for x,y in df['geometry'].iteritems():
    if len(y) > 1:
        for i in y:
            feature = {
                    'type': 'Feature',
                    'id': x,
                    'geometry':
                    {'type': 'Polygon',
                    'coordinates': i},
                    'properties':
                    {"description": df['cn'].loc[x],
                    "fill": "#22016e",
                    "fill-opacity": 0.4,
                    "stroke": "#ffffff",
                    "stroke-width": "1",
                    "stroke-opacity": 1}}
            features_list.append(feature)
    if len(y) == 1:
        for i in y:
            feature = {
                    'type': 'Feature',
                    'id': x,
                    'geometry':
                    {'type': 'Polygon',
                    'coordinates': i},
                    'properties':
                    {"description": df['cn'].loc[x],
                    "fill": "#22016e",
                    "fill-opacity": 1,
                    "stroke": "#ffffff",
                    "stroke-width": "1",
                    "stroke-opacity": 1}}
            features_list.append(feature)
```

```python
geo['features'] = features_list
with open('result2.json', 'w') as fp:
    json.dump(geo, fp)
```
